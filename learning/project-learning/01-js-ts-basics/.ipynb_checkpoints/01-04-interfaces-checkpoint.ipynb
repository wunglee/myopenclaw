# 1.4 接口（interface）

> 目标：理解 `interface` 的用法，以及与 `type` 的区别

## 什么是接口？

`interface` 用于定义对象的结构，和 `type` 类似但有扩展能力。

```typescript
// 定义接口
interface GatewayConfig {
  port: number;
  host: string;
  verbose?: boolean;
}

// 实现接口
const config: GatewayConfig = {
  port: 8080,
  host: "localhost",
  verbose: true
};

// 函数使用接口作为参数类型
function startGateway(cfg: GatewayConfig): string {
  return `启动网关: ${cfg.host}:${cfg.port}`;
}

console.log(startGateway(config));
```

## type vs interface：什么时候用哪个？

| 特性 | type | interface |
|------|--------|-------------|
| 定义对象结构 | 支持 | 支持 |
| 联合类型 | 支持 | 不支持 |
| 扩展 | 不支持 | 支持 extends |
| 重复定义 | 报错 | 自动合并 |

建议：简单对象用 type，需要扩展用 interface。
