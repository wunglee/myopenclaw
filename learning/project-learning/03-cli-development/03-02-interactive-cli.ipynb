# 03-02 交互式 CLI

使用 inquirer 和 @clack/prompts 创建交互式命令行界面。

## 1. Inquirer - 经典交互库

```typescript
import inquirer from 'inquirer';

const answers = await inquirer.prompt([
  {
    type: 'input',
    name: 'projectName',
    message: '项目名称:',
    default: 'my-project',
    validate: (input) => input.length > 0 || '名称不能为空'
  },
  {
    type: 'list',
    name: 'framework',
    message: '选择框架:',
    choices: ['Express', 'Fastify', 'Koa']
  },
  {
    type: 'confirm',
    name: 'install',
    message: '立即安装依赖?',
    default: true
  }
]);

console.log('配置:', answers);
```

## 2. @clack/prompts - 现代美观

```typescript
import * as p from '@clack/prompts';

p.intro('创建新项目');

const project = await p.group({
  name: () => p.text({
    message: '项目名称?',
    placeholder: 'my-app'
  }),
  type: () => p.select({
    message: '项目类型?',
    options: [
      { value: 'api', label: 'API 服务' },
      { value: 'cli', label: 'CLI 工具' }
    ]
  })
});

p.outro('项目创建成功');
```
