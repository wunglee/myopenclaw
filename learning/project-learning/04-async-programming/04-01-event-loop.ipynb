# 04-01 事件循环 (Event Loop)

理解 Node.js 事件循环机制，这是异步编程的核心。

## 什么是事件循环？

Node.js 使用事件驱动模型：

1. 所有 I/O 操作都是异步的
2. 操作完成后通过回调通知
3. 事件循环协调这些回调的执行

```typescript
// 演示执行顺序
console.log('1. 同步代码开始');

setTimeout(() => {
  console.log('4. setTimeout (宏任务)');
}, 0);

Promise.resolve().then(() => {
  console.log('3. Promise (微任务)');
});

console.log('2. 同步代码结束');

// 输出顺序：1 -> 2 -> 3 -> 4
```

## 宏任务 vs 微任务

| 类型 | 示例 | 优先级 |
|------|------|--------|
| 微任务 | Promise, queueMicrotask | 高 |
| 宏任务 | setTimeout, setInterval | 低 |

```typescript
// 微任务优先
setTimeout(() => console.log('timeout'), 0);

Promise.resolve().then(() => console.log('promise'));

queueMicrotask(() => console.log('microtask'));

// 输出：promise -> microtask -> timeout
```
