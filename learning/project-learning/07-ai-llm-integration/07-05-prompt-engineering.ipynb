```typescript
const systemPrompt = `You are OpenClaw Assistant, a helpful AI for managing messaging platforms.

Core capabilities:
- Send and receive messages across multiple platforms
- Manage conversation contexts
- Execute approved actions

Rules:
1. Always confirm before executing destructive actions
2. Maintain professional tone
3. Use available tools when appropriate
4. If unsure, ask for clarification

Current context: {{context}}`;
```

## 2. Few-Shot 提示

```typescript
const fewShotPrompt = `Extract the intent from the user message.

Examples:
User: "Send a message to John"
Intent: {"action": "send_message", "recipient": "John", "content": null}

User: "What's the weather in Beijing?"
Intent: {"action": "get_weather", "location": "Beijing"}

User: {{userMessage}}
Intent:`;
```

## 3. Chain-of-Thought

```typescript
const cotPrompt = `Solve this step by step:

Question: {{question}}

Let's think through this:
1. First, identify the key information...
2. Then, analyze the relationships...
3. Finally, formulate the answer...

Answer:`;

// 使用 ReAct 模式
const reactPrompt = `You are an AI assistant. Use the following format:

Thought: Consider what to do
Action: The action to take (one of: [tools])
Action Input: The input to the action
Observation: The result of the action
... (repeat as needed)
Thought: I now know the final answer
Final Answer: The final answer`;
```
